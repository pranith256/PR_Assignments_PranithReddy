## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`.

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

2. Use `numpy.random.shuffle` to shuffle the indices. Call the shuffled array, `rows`

4. Print `rows` and inspect the elements.

```python
# Numpy array that contains number in range(0 to size of avg_img)
rows = np.arange(n)
#Using random.shuffle to shuffle the indices.
np.random.shuffle(rows)
#Printing shuffled rows
print(rows)
#rows contain the shuffled order of row_indices.
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
#creating a T matrix by shuffling the rows indices in identity matrix
T = identity_matrix[rows]
print(T)
#What is matrix T
'''
T is one of the permutation of the identity matrix. The rows in the identity matrix are shuffled to form T.
The elements in the T matrix are same as Identity_matrix, but rows are shuffled corresponding to rows array
'''
#Sanity check
print("Shape of T matrix", T.shape)
#Check the size and reshape if needed.
if T.shape != identity_matrix.shape:
    T = T.reshape(identity_matrix.shape)
```

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see.
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image.

```python
#Multiply T and img_average_channel
transformed_image = np.dot(T, img_average_channel)
#plotting transformed_image
print(transformed_image)
print("Shape of transformed_image:", transformed_image.shape)

#pattern observation
'''
As the img_average_channel is multiplied with the T(which is a shuffling of the rows of identity matrix) it also rearranges the order of the img_average_channel

This dot product is making changes in spatial relationships of the original image's row and will alter visual affects
depending on the shuffling order of rows.
This is a way of altering the structure of the image, where pixel values are rearranges with sae values.
'''
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
#we will find the determinant of T and check if it is non-zero matrix. If it is non-zero then
#it is invertible else not.
#Compute determinant of T
det_T = np.linalg.det(T)
# Check if 'T' is invertible (determinant is nonzero)
if det_T != 0:
    # Compute the inverse of 'T'
    Tinv = np.linalg.inv(T)
else:
    print("Matrix T is not invertible.")

#Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.
original_image = np.dot(Tinv, transformed_image)
original_image = np.round(original_image).astype(np.uint8)
#Plotting the original_image
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')
plt.show()
```

## Part 8: Extra Points

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image.

```python
#Shuffling the column indices
columns = np.arange(n)
np.random.shuffle(columns)
#creating a T matrix using the columns rearrangement.
T_columns = np.eye(n)[:,columns]
#Multiplying the original image with the T matrix.
scrambled_image = np.dot(original_image, T_columns)
print("Shape of Scrambled Image:", scrambled_image.shape)
#Plotting the scrambled image.(proof of concept)
plt.imshow(scrambled_image, cmap='gray')
plt.title('Scrambled Image')
plt.show()
```

```python
#now lets perform scramble on both rows and columns
n_rows = np.arange(n)
n_columns = np.arange(n)
np.random.shuffle(n_rows)
np.random.shuffle(n_columns)

#Creating T matrix for both rows and columns shuffling
T_rows = np.eye(n)[n_rows, :]
T_columns = np.eye(n)[:, n_columns]

#scrambled image of both the T_row and T_column
scrambled_image = np.dot(np.dot(T_rows, original_image), T_columns)
#Shape of the image.
print("Shape of Scrambled Image:", scrambled_image.shape)
#Proof of concept for rows,columns shuffling
plt.imshow(scrambled_image, cmap='gray')
plt.title('Scrambled Image')
plt.show()
```

```python
#Inverting the operations of T matrix and retieving the original image.
original_image_1 = recovered_image = np.dot(np.dot(np.linalg.inv(T_rows), scrambled_image), np.linalg.inv(T_columns))
#Plotting the original image
plt.imshow(recovered_image, cmap='gray')
plt.title('Recovered Original Image')
plt.show()
```
